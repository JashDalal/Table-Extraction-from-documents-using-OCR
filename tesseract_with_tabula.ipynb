```python
# tabula-py is just a wrapper around tabula-java
ppm_path = r'images/ppm/'
img_path = r'images/sample/page1'
img_folder = r'images/sample/'
pdf_path = r"pdfs/working/sample_table_data.pdf"
output_path = r"csv files/split.csv"
```

```python
import PIL
from PIL import Image
import cv2
import pytesseract
import img2pdf
from pdf2image import convert_from_path
import tabula

pytesseract.pytesseract.tesseract_cmd = r"C:\Program Files\Tesseract-OCR\tesseract.exe"
def tesseract_convert(img_file):
    img_path = img_folder + img_file
    with Image.open(img_path) as img:
        tesseract_output = pytesseract.image_to_string(img, lang='eng', config='--psm 3')
        img.close()
    return tesseract_output
```

```python
pdf_pages = 0
pages = convert_from_path(pdf_path, output_folder = ppm_path)
pdf_file = (pdf_path.split('/')[-1])[:-4]

for page in pages:
    page.save(img_folder +"page%d.jpg" % pages.index(page), "JPEG")
    pdf_pages += 1
```

```python
output = []
for i in range(pdf_pages):
    img_file = "page%d"%i + ".jpg"
    output.append(tesseract_convert(img_file))
```

```python
output = '\n'.join(output)
```
